```python
import numpy as np
x = np.arange(0,100)
y = x*2
z=x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, 'r--')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Titulo")
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

ax.plot(x,y, 'r')
ax2.plot(x,y, 'r')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

ax.plot(x,z, 'r')
ax2.plot(x,y, 'r')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, color="blue", ls='--', lw=3)
axes[1].plot(x, z, color="red", ls='-', lw=3)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 2))
axes[0].plot(x, y, color="blue", ls='-', lw=3)
axes[1].plot(x, z, color="red", ls='--', lw=3)
```
