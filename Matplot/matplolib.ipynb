```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
y = x*x
```

```python
plt.plot(x, y, 'r--')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Titulo')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g.-')
```

```python
fig = plt.figure()
axe1 =  fig.add_axes([0.1, 0.1, 0.8, 0.8])
axe1.plot(x, y)
axe1.set_xlabel('Eixo X')
axe1.set_title('Axe 1')

axe2 =  fig.add_axes([0.2, 0.5, 0.3, 0.3])
axe2.plot(y, x)
axe2.set_ylabel('Eixo Y')
axe2.set_title('Axe 2 ')
```

```python
fi, ax = plt.subplots()
ax.plot(x, x**3, 'b--')
ax.plot(x, x**4, 'g--')
ax.set_xlabel("Eixo x")
ax.set_ylabel("Eixo y")
ax.set_title("Titulo")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].plot(x, x**3, 'b--')
ax[0].set_xlabel("Eixo x")
ax[0].set_ylabel("Eixo y")
ax[0].set_title("Titulo")
ax[1].plot(x, x**4, 'g--')
ax[1].set_xlabel("Eixo x")
ax[1].set_label("Eixo y")
ax[1].set_title("Titulo")
```

```python
fig, ax = plt.subplots(nrows=5, ncols=5)
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(12,8))
axes.plot(x, y, 'r--', label='x^2')
axes.plot(y, x, 'g--', label='log x (y)')
axes.legend(loc=0)
axes.set_title("titulo")
axes.set_xlim([0,2])
axes.set_ylim([0,10])
```
