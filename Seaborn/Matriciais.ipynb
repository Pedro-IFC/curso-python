```python
import seaborn as sns
import numpy as np
import scipy
%matplotlib inline
```

```python
flights = sns.load_dataset('flights')
tips = sns.load_dataset('tips')
```

```python
sns.heatmap(tips.select_dtypes(include=[np.number]).corr(), cmap='coolwarm', annot=True)
```

```python
pf = flights.pivot_table(values='passengers', index='month', columns='year', observed=False)
```

```python
sns.heatmap(pf, cmap='coolwarm', annot=False)
```

```python
sns.clustermap(pf)
```
